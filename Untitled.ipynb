```python
import pandas as pd

glassdoor = pd.read_csv('scraped_from_glassdoor.csv', sep = ';')

print('Data points: {}'.format(len(glassdoor)))

glassdoor['date'] = pd.to_datetime(glassdoor['date'], format = '%B %d, %Y')
glassdoor = glassdoor[glassdoor['date'].notna()]
print('Data points after removing NaN: {}'.format(len(glassdoor)))


pre_metoo = glassdoor[glassdoor['date']<='2017-10-31'].copy()
post_metoo = glassdoor[glassdoor['date']>'2017-10-31'].copy()

print('Pre #MeToo: {}'.format(len(pre_metoo)))
print('Post #MeToo: {}'.format(len(post_metoo)))

post_metoo['cons']
```

```python
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


stop_words = set(stopwords.words('english'))

text = ' '.join(pre_metoo['cons'])
word_tokens = word_tokenize(text)
filtered_text = ' '.join([w for w in word_tokens if not w.lower() in stop_words])

# Create and generate a word cloud image:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white",width=800, height=400).generate(filtered_text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

text = ' '.join(post_metoo['cons'])
word_tokens = word_tokenize(text)
filtered_text = ' '.join([w for w in word_tokens if not w.lower() in stop_words])

# Create and generate a word cloud image:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white",width=800, height=400).generate(filtered_text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()


```
